```python

#營業收入func(營業毛利率/營業利益率/稅前淨利率) 回傳json
def getOperatingRevenue(company):
    d = {}
    if '鴻海' in company or '2317' in company: 
        df = r.Report_TW2317()
    elif '台積電' in company or 'tsmc' in company or '2330' in company:
        df = r.Report_TW2317()
    else:
        d['hasData'] = False
        return d
    #exist data
    df = df[0:3] #只取前3個 營業毛利率/營業利益率/稅前淨利率
    
    d['hasData'] = True
    d['columns'] = df.columns
    d['data'] = df.values
    return d

getOperatingRevenue('2317')
```
